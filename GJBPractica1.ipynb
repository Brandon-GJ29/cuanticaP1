# 1.Ejercicio 1.

Crea un circuito cuántico con un solo qubit y aplícale 5 compuertas diferentes.


### 1.1. Crear el circuito. 

```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
#Primero se crea un circuito con un solo cubit
circ1 = QuantumCircuit(1)    
# Se obtiene el estado después de crear el circuito
psi_0 = Statevector(circ1) 
```

### 1.2. Aplicar Compuerta H

```python

# Se aplica la compuerta H al estado |0>, obteniedo  |+>
circ1.h(0)                   
# Se obtiene el estado después de aplicar H
psi_1 = Statevector(circ1)   
display(circ1.draw('mpl'))   
```

### 1.3. Aplicar Commpuerta Z

```python
# Aplicar compuerta Z, Obteniendo  el estado |->
circ1.z(0)                  
# Se obtiene el estado después de aplicar Z
psi_2 = Statevector(circ1)
display(circ1.draw('mpl'))  
```

### 1.4. Aplicar Compuerta H

```python
# Se Aplica la compuerta H de nuevo, quedamos en el estado |1>
circ1.h(0)                   
# Se obtiene el estado después de aplicar H
psi_3 = Statevector(circ1)   
display(circ1.draw('mpl')) 
```

### 1.5. Aplicar Compuerta X

```python
# Se Aplica la compuerta X , quedamos en el estado i∣+>
circ1.x(0)  
psi_4 = Statevector(circ1) 
display(circ1.draw('mpl')) 
```

### 1.6. Aplicar segunda compuerta X.

```python
# Se Aplica la compuerta X , quedamos en el estado 
circ1.x(0)  
psi_5 = Statevector(circ1)  
display(circ1.draw('mpl'))   # mostrar el circuito
```
